## Auto MPG Dataset

```python
#importing packages
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading dataset
auto = pd.read_csv(r"D:\Dataset\auto+mpg\auto-mpg.data",header=None,delim_whitespace=True)
auto
```

```python
#renaming the columns
auto.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name']
auto.head()
```

```python
#Check missing values
auto.isnull().sum()
```

```python
#Check datatypes
auto.dtypes
```

```python
#Check unique values
for i in auto.columns:
    print(i,':','\n',auto[i].unique(),'\n')
```

```python
for i in auto.columns:
    print(i,':',sum(auto[i]=='?'))
```

```python
d = auto.describe(include='all')
d
```

```python
auto.horsepower.replace('?',d['horsepower'][2],inplace=True) #auto.horsepower = auto.horsepower.replace('?',d['horsepower'][2])
```
