## Matplotlib

Data visualization Technique

```python
import matplotlib.pyplot as plt
```

```python
#Line Plot
x = [3,6,9,4,7,6,10]
y = [5,9,10,2,4,5,8]

plt.plot(x,y,'o--')
plt.plot(x,y,'ro')
plt.title('Line Plot')
plt.xlabel('X - Axis',color='g')
plt.ylabel('Y - Axis',color='purple')
plt.grid()
plt.show()
```

```python
#Bar Plot
x = [1,3,5,7,9]
y = [2,4,6,8,10]

plt.bar(x,y)
plt.title('Bar Plot')
plt.xlabel('X - Axis',color='g')
plt.ylabel('Y - Axis',color='purple')
#plt.grid()
plt.show()
```

```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]
plt.bar(x,y)
plt.show()

x1 = [4,7,2,9,8]
y1 = [9,2,6,4,1]
plt.bar(x1,y1)
plt.show()

plt.title('Bar Plot')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
#plt.grid()
plt.show()
```

```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]
plt.bar(x,y,label='Example 1')


x1 = [4,7,2,9,8]
y1 = [9,2,6,4,1]
plt.bar(x1,y1,label='Example 2')


plt.title('Bar Plot')
plt.legend(loc='lower center')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
#plt.grid()
plt.show()
```

```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]
plt.bar(x,y,label='Example 1')
plt.show()

x1 = [4,7,2,9,8]
y1 = [9,2,6,4,1]
plt.bar(x1,y1,label='Example 2')
plt.show()

plt.title('Bar Plot')
plt.legend(loc='lower center')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
#plt.grid()
plt.show()
```

```python
#Histogram
ages = [2,10,5,6,4,9,3,4,35,28,21,45]
bins = [2,10,20,30,40,50]

plt.hist(ages,bins,histtype='bar',rwidth=0.5,color='orange',edgecolor='k')
plt.title('Histogram')
plt.show()
```

```python
#Scatter plot
x = [3,5,4,7,2,10,6,9]
y = [2,6,4,9,6,9,3,10]
plt.scatter(x,y,s=2000,color='orange',marker='1')
plt.title('Scatter Plot')
plt.show()
```

```python
x = [3,5,4,7,2,10,6,9]
y = [2,6,4,9,6,9,3,10]
plt.bar(x,y,color='g',width=0.05)
plt.scatter(x,y,s=300,color='orange',marker='*',edgecolor='orange',linewidth=12)
plt.scatter(x,y,color='brown')


plt.title('Scatter Plot')
plt.show()
```

```python
#Pie chart
slices = [10,4,8,3,5]
activities = ['Phone','Eating','Sleeping','Studying','Playing']
colors = ['b','gold','r','m','g']

plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0,0,1,0),autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv(r"D:\Dataset\iris\iris.data")
iris
```

```python
iris.columns = ['SL','SW','PL','PW','Flower']
iris.head(3)
```

```python
#Distribution Plot
sns.distplot(iris.SL)
plt.show()
```

```python
sns.histplot(iris.SL)
plt.show()
```

```python
sns.kdeplot(iris.SL)
plt.show()
```

```python
sns.histplot(iris.SL,fill=False,color='purple',hatch='*')
plt.show()
```

```python
sns.kdeplot(iris.SL,fill=True,color='c',hatch='/')
plt.show()
```

```python
sns.distplot(iris.SW[iris['Flower']=='Iris-setosa'])
sns.distplot(iris.SW[iris.Flower=='Iris-versicolor'])
sns.distplot(iris.SW[iris.Flower=='Iris-virginica'])
plt.show()
```

```python
sns.histplot(iris.SW[iris['Flower']=='Iris-setosa'])
sns.histplot(iris.SW[iris.Flower=='Iris-versicolor'])
sns.histplot(iris.SW[iris.Flower=='Iris-virginica'])
plt.show()
```

```python
sns.displot(iris.SW[iris['Flower']=='Iris-setosa'])
sns.displot(iris.SW[iris.Flower=='Iris-versicolor'])
sns.displot(iris.SW[iris.Flower=='Iris-virginica'])
plt.show()
```

```python
sns.kdeplot(iris.SW[iris['Flower']=='Iris-setosa'],fill=True)
sns.kdeplot(iris.SW[iris.Flower=='Iris-versicolor'],fill=True)
sns.kdeplot(iris.SW[iris.Flower=='Iris-virginica'],fill=True)
plt.show()
```
